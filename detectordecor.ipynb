<h2>Descrição do projeto</h2>
<p style="text-align: justify;">   O projeto consiste na realização de um sistema de detecção e identificação de cores em tempo real, utilizando visão computacional e a linguagem Python.<br>
Por meio da câmera do dispositivo do usuário, o sistema captura imagens continuamente e processa esses frames em tempo real. A partir das imagens capturadas, o projeto analisa uma área centralisada, na qual os valores das cores são convertidos para o modelo HSV (Hue, Saturation, Value), o que possibilita uma identificação mais precisa das cores, mesmo em diferentes iluminações.<br>
  O código em Python faz a média dos valores capturados na região delimitada e mostra para o usuario a cor predominante, identificando as seguintes tonalinades: vermelho, laranja, amarelo, verde, ciano, azul, roxo e rosa e as variações: branco, preto e cinza. <br>
  Para melhor visualização da cor detectada foi implementado um delay (mecanismo de atraso), evitando oscilações rápidas nas mudanças de cores.<p>



<h2>O que é o HSV?</h2>
<p style="text-align: justify;">A sigla HSV (Hue, Saturation, Value) significa, em português, Matiz, Saturação e Valor, e é um modelo de representação de cores semelhante à percepção humana.<br>
O HSV separa as cores em três componentes: Hue (matiz), que representa a cor base; Saturation (saturação), que representa a intensidade da cor; e Value (valor), que representa o brilho da cor.<br>
Abaixo é possível observar uma representação do modelo HSV.</p>

<h3>Definição da função mantem_proporcao</h3>
<p style="text-align: justify;">Essa função recebe dois parametros: frame (a image em si) e target (o tamanho desejado).<br>
Ela calcula o quanto é neccessário multiplicar a largura original da para chegar no tamanho desejado (target).<br>
Logo após o cálculo, aplica esse mesmo fator à altura da imagem, garantindo que a proporção seja mantida.<br>
Por último, a imagem é redimencionada de acorodo com os novos valores.
Ela é posteriormente usada no while, especificadamente na linha de código:  
    <code>frame_invertido = mantem_proporcao(frame_invertido)</code>
</p>

```python
def mantem_proporcao(frame, target=1300):
    h, w = frame.shape[:2]
    scale = target / w
    target_h = int(h * scale)
    return cv2.resize(frame, (target, target_h))
```

<h2>Bibliotecas</h2>
<p style="text-align: justify;">O sistema utiliza algumas bibliotecas que precisam ser importadas antes de serem utilizadas.<br>
    
<strong>cv2: </strong>Utilizada para processamento de imagens e vídeos. Permite capturar as imagens, manipular frames, converter cores e exibir informaçoes na tela do usuário.<br>

<strong>numpy, apelidado de np: </strong>Utilizada para manipulação de matrizes. No projeto é usada para trabalhar com os dados das imagens.<br>

<strong>time: </strong>Utilizada para controle do tempo e para gerar um pequeno delay, que promove estabilidade na detecção das cores. </p>

```python
import cv2
import numpy as np
import time
```

<h2>Definição da função identifica_cor_hsv</h2>
<p style="text-align: justify;">Essa função recebe três parâmentros que são utilizados para classificar as cores : 
    
<ol>
  <li>h (Hue / Matiz) que representa o tipo da cor </li>
  <li>s (Saturation / Saturação) que indica a intensidade da cor</li>
  <li>v(Value / Valor) que representa o brilho da cor</li>
</ol>

Dentro dessa função, exixtem várias condições que ajudam nessa classificação. Por exemplo,, se o brilho(v) for menor que 50 o sistema irá dizer que a cor é preta, independente de outros valores, se a matiz(h) for menor que 175 mas maior que 155 o sistema irá identificar a cor como rosa.<br>
Além disso, existe um caso padrão: se o sistema identificar uma cor que não está dentro dos parâmentros pré-estabelicidos, ele retorna a cor vermelho.<br>

Ela é posteriormente usada dentro do while, especificadamente na linha de código:
<code>cor = identifica_cor_hsv(media_h, media_s, media_v)</code>
</p>

```python
def identifica_cor_hsv(h, s, v):

    if v < 50:
        return "preto"

    if s < 25:
        if v > 200:
            return "branco"
        elif v > 80:
            return "cinza"
        else:
            return "preto"

    if h < 5 or h >= 175:
        return "vermelho"
    elif h < 25:
        return "laranja"
    elif h < 35:
        return "amarelo"
    elif h < 80:
        return "verde"
    elif h < 100:
        return "ciano"
    elif h < 130:
        return "azul"
    elif h < 155:
        return "roxo"
    elif h < 175:
        return "rosa"
    else:
        return "vermelho"
```

<h2>Inicialização da câmera</h2>
<p style="text-align: justify;">Esse trecho é responsável por inicializar a câmera e verificar se ela foi aberta corretamente.<br>
    Ele cria um objeto que inicializa a captura do vídeo, usando a câmera do dispositivo. O número 0 indica que a câmera que será usada é a padrão(normalmente a do própio dispositivo).<br>
    A função <code>isOpened()</code> verifica se a câmera foi aberta corretamente e retorna <code>True</code> ou <code>False</code>. Se retornou <code>False</code> o sistema é fechado automáticamente e se retornou <code>True</code> o sistema continua a rodando normalmente.
</p>

```python
camera = cv2.VideoCapture(0)

if not camera.isOpened():
    print("Não conseguiu abrir a câmera")
    exit()
```

<h2>Definindo variáveis de controle</h2>
<p style="text-align: justify;">Essa trecho contém variáveis que definem a cor atual que o sistema esá identificando, a última cor detectada, o tempo da última mudança e valor do delay. <br>
    Todas essas variáveis serão posteriormente usadas para garantir maior estabilidade na identificação da cor e na exibição dos resultados.</p>

```python
rodando = True
cor_atual = ""
ultima_cor = ""
tempo_ultima_mudanca = 0
delay = 2.0  
```

<h2>Loop principal</h2>
<h4>Inicialização do programa</h4>
<p style="text-align: justify;"> O comando <code>while</code> é responsável por iniciar um loop de repetição, que mantém o sistema em execução enquanto a variável <code>rodando</code> for verdadeira.  <br>
Dentro do loop, o sistema tenta capturar um frame da câmera por meio da função <code>camera.read()</code>, que retorna um indicador de sucesso e o frame capturado.<br>
Caso a captura não seja bem-sucedida, o sistema exibe uma mensagem de erro para o usuário e interrompe o loop com o comando <code>break</code>, encerrando a execução do sistema.<br>
Se a câmera estiver funcionando corretamente, o sistema continua rodando normalmente, processando os frames capturados.<br><br>
</p>
<h4>Tratamento e exibição da imagem.</h4>
<p style="text-align: justify;">  O código <code>frame_invertido=cv2.flip(frame,1)</code> inverte o frame capturado horizontalmente, proporcionando uma melhor experiência ao usuário(efeito espelho).<br>
Em seguida, são definidos os valores <code>crop_w</code> e o <code>crop_h</code>, que serão usados para recortar as bordas da imagem.<br>
 A função <code>mantem_proporcao</code> ajusta o tamanho do frame mantendo sua proporção. Em seguida, o sistema recorta a imagem com base nas margens definidas.<br>
 Após isso, o sistema obtém a altura e a largura da imagem, calcula o ponto central(x e y) e define uma área ao redor desse ponto, que será usada para a leitura da cor.<br>
Por fim, são utilizados <code>min </code> e <code>max </code> para garantir que essa área central não saia da tela
<br><br>
</p>
<h4>Identificação da cor</h4>
<p style="text-align: justify;"> 
Através do código: <code> hsv = cv2.cvtColor(frame_cortado, cv2.COLOR_BGR2HSV)</code>, o sistema converte as cores BGR (padrão do OpenCV) para o modelo HSV, facilitando a identificação das cores.<br>
    É feita a delimitação da região de interesse utilizando as coordenadas previamente definidas (<code>y1:y2</code> e <code>x1:x2</code>). Esse será o local onde a cor será analisada.<br>
O sistema faz uma média dos valores de matiz, saturação e valor (HSV) obtidos em toda a região, utilizando a função <code>np.mean</code>. Essa média faz com que a cor obtida seja mais precisa e estável.<br>
Com os valores obtidos, a função <code>identifica_cor_hsv</code> é chamada para determinar o nome da cor predominante.<br>
Após armazenar o nome da cor, o sistema utiliza o <code>time.time()</code> para controlar a estabilidade da exibição. Caso a cor identificada seja diferente da última, o sistema só irá exibir a nova cor após um tempo definido (<code>delay</code>). Caso a cor seja a mesma, ela é mantida.
</p>
<br>
<h4>Exibição de resultados na tela.</h4>
<p style="text-align: justify;"> A função <code>cv2.putText</code> é usada para escrever informações no frame exibido.
O código exibe primeiramente o nome da cor detectada através do código: <code>cv2.putText(frame_cortado, cor, (10, 50), cv2.FONT_HERSHEY_SIMPLEX,1, (0, 0, 255), 2)</code><br>
Em seguida, é criada uma string contendo os valores HSV (<code>H</code>, <code>S</code> e <code>V</code>) que também é exibida para o usuário, permitindo identificar os valores numéricos da cor identificada.<br>
O sistema também emite uma instrução de como sair do programa. <br>

O código também desenha elementos gráficos: <br>
<ol>
<li>Um retângulo, usando o código <code>cv2.rectangle(frame_cortado, (x1, y1), (x2, y2), (0, 255, 0), 2)</code>, que indica a região analisada.<br></li>
<li>Um círculo, usando o código <code>cv2.circle(frame_cortado, (x, y), 5, (0, 0, 255), -1)</code>, que indica o ponto central da leitura.<br></li>
</ol>
A função <code>cv2.imshow</code> é responsável por exibir o frame com todas essas informações na tela do usuário.<br>
O sistema verifica se o usuário clicou em 'q' através do  comando <code>cv2.waitKey(1)</code>. Caso tenha apertado, o loop é interrompido e o sistema é fechado.
    <br><br>
    </p>
<h4>Finalização e liberação de recursos.</h4>
<p style="text-align: justify;">Ao encerrar o sistema, o código <code>camera.release()</code> libera a câmera de uso e fecha todas as janelas pelo código <code>cv2.destroyAllWindows()</code>. Essas ações garantem que todos os recursos sejam encerrados definitivamente e que a câmera possa ser usada por outro programa depois desse ser encerrado. </p>

```python
while rodando:
    status, frame = camera.read()

    if not status:
        print("Erro ao acessar a câmera")
        break
    
    # inverte a camera
    frame_invertido=cv2.flip(frame,1)
    crop_w = 160
    crop_h = 90
         # coloca uma medida exata para a janela
    
    frame_invertido = mantem_proporcao(frame_invertido)
    frame_cortado = frame_invertido[crop_h:-crop_h, crop_w:-crop_w]

    altura, largura, _ = frame_cortado.shape

    x = largura // 2
    y = altura // 2

    tamanho = 25  # tamanho da área de leitura

    # proteger contra sair da tela
    y1 = max(0, y - tamanho)
    y2 = min(altura, y + tamanho)
    x1 = max(0, x - tamanho)
    x2 = min(largura, x + tamanho)

    # converter para HSV
    hsv = cv2.cvtColor(frame_cortado, cv2.COLOR_BGR2HSV)

    # recorte da região
    regiao = hsv[y1:y2, x1:x2]

    # média HSV
    media_h = int(np.mean(regiao[:, :, 0]))
    media_s = int(np.mean(regiao[:, :, 1]))
    media_v = int(np.mean(regiao[:, :, 2]))

    # identificar cor
    cor = identifica_cor_hsv(media_h, media_s, media_v)
    agora = time.time()
    # faz com que a cor mostrada na tela fique por um certo tempo e não oscilando
    if cor != ultima_cor:
        if agora - tempo_ultima_mudanca > delay:
            cor_atual = cor
            ultima_cor = cor
            tempo_ultima_mudanca = agora
    else:
        cor_atual = cor

    # mostrar nome da cor
    cv2.putText(frame_cortado, cor, (10, 50),
                cv2.FONT_HERSHEY_SIMPLEX,
                1, (0, 0, 255), 2)

    # mostrar HSV
    texto_hsv = f"H:{media_h} S:{media_s} V:{media_v}"
    cv2.putText(frame_cortado, texto_hsv, (10, 100),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.7, (0, 0, 255), 2)

    # instrução sair
    cv2.putText(frame_cortado, "Aperte q para sair", (10, 140),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.7, (0, 0, 255), 2)

    # desenhar área analisada
    cv2.rectangle(frame_cortado, (x1, y1), (x2, y2), (0, 255, 0), 2)

    # ponto central
    cv2.circle(frame_cortado, (x, y), 5, (0, 0, 255), -1)

    cv2.imshow("Detector de Cor (HSV)", frame_cortado)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

camera.release()
cv2.destroyAllWindows()
```
